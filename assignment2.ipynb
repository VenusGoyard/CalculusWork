##Problem 4: Normal Distribution

Use the information given about the mean and standard deviation of a normal distribution to answer the questions.

- Suppose 𝜇=10,𝜎=3. Draw a plot of this distribution on 𝑥=[0,20]. What is the probability 𝑥>15?



```python
x = np.linspace(0,20,200)
 
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density

mean = 10
sd = 3

pdf = normal_dist(x,mean,sd)
 
plt.plot(x,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```

Answer: The probability that x > 15 is 0.2.
